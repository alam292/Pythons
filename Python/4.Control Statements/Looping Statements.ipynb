#### Iteration / Looping Statements :-

Used to repeat a block of code multiple times.

##### for loop :-

Iterates over a sequence (list,string ,range)

```python
a = [1,2,3,4,5]
for x in a:
    print(a)
```

```python
a = 'apple'
for x in a:
    print(a)
```

```python
for i in range (1,6):
    print(i)
```

```python
for x in "banana":
    print(x)
```

```python
a = "banana"
for x in a:
    print(x)
```

##### while loop :-

Runs while condition is true.

```python
i = 1
while i<=5 :
    print(i)
    i+=1
```

##### Nested loop :-

Loop inside another loop.

```python
for i in range(1,4):
    for j in range(1,4):
        print(i,j)
```
