#### Control Statements :-

Control statements decide how your program runs step-by-step.

They control conditions, loops, and flow jumps.

#### Types :-
| Type     | Statements                      |
| -------- | ------------------------------- |
| Decision | `if`, `if-else`, `if-elif-else` |
| Looping  | `for`, `while`                  |
| Jump     | `break`, `continue`, `pass`     |
