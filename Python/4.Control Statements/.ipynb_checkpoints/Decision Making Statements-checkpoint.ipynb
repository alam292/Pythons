#### Decision Making Statements :-

Used when you want to take decision based on conditions.

Works using conditions (True/False).



##### if statements :- 

Executes block if condition is true.

```python
age = 45
if age >= 37:
    print('Eligible to vote')
```

##### if-else statements:-

Executes one block if true, another if false.

```python
num = 5
if num%2 == 0:
    print('Even')
else:
    print('Odd')
```

##### if-elif-else statements (ladder) :-

Used for multiple condition.

```python
marks = 78
if marks >= 90:
    print("Grade A")
elif marks >= 70:
    print("Grade B")
elif marks >= 50:
    print("Grade C")
else:
    print('Fail')
```

##### Nested if :-

if inside another if.

```python
num = 10
if num > 0:
    if num % 2 == 0:
        print("Positive Even")
```
