#### Jump / control transfer Statements :-

Used to change the normal flow of loops.

##### break :-

Terminates  the loop completely.

```python
for i in range (1,10):
    if i == 5:
        break
    print(i)
```

##### continue :-

Skips current iteration.

```python
for i in range(1,6):
    if i == 3:
        continue
    print(i)
```

##### pass :-

Does nothing (placeholder).

```python
for i in range(5):
    pass
```
