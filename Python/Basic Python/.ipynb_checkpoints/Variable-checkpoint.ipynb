### Variable :- 

A variable in Python is a container for storing data.

It is a named memory location that holds values.

### Rules :-

1.Can contain letters (A-Z, a-z) , digits (0-9) and underscore(_).

2.Must start with a letter or an underscore(_).

3.Can't start with a digits.

4.Can't use python keywords as variable names.

5.Case-Sensitive (Name , name are different variable).

```python
name = 'Jon'
_age = 28
user123 = 'Alice'
```

#### Multi words variable names

##### camel case :- Each word , except the first , starts with a capital letter.

```python
myVariableName = 'Jon'
```

##### pascal case :- Each word starts with a capital letter.

```python
MyVariableName = 'Jon'
```

##### snake case :- Each word is separted by an underscore character.

```python
my_variable_name = 'jon'
```

#### Case-Sensitive

```python
x = 'john'
```

```python
X = 4
```

#### Assign mutliple values

```python
x,y,z = 'orange','mango','apple'
print(x)
print(y)
print(z)
```

```python
x,y,z = 'orange','mango','apple'
print(y)
print(z)
print(x)
```

#### one value to multiple variable 

```python
x = y = z = 'orange'
print(x)
print(y)
print(z)
```

#### unpack collection :-
A collection of values in a list, tuple etc .Python allows you to extract the values into variables. This is called unpacking.

```python
fruits = ['apple','orange','mango']
x,y,z = fruits
print(x)
print(y)
print(z)
```

#### Swaping two variables

```python
a, b = 5 , 10
print(a,b)
print('Swaping')
a, b = b, a
print(a, b)
```
