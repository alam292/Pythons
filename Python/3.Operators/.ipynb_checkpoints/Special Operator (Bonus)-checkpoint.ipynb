#### Ternary operator :- 

The ternary operator is one-line shortcut for if-else.

```python
a = 10
b = 20

val = a if a>b else b
print(val)
```

```python
n = int(input('Enter a number :'))
even_odd = 'Even' if n%2==0 else 'Odd'
print(even_odd)
```

##### Nested Ternary 

```python
num = 0
result = 'Positive' if num > 0 else "Negative" if num <0 else "Zero"
print(result)
```
