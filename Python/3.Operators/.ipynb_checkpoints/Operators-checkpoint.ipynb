### Operators :- 

Operators are used to perform operations on variables and values.

##### Types :-

| Type       | Description        | Example        |    |
| ---------- | ------------------ | -------------- | -- |
| Arithmetic | Math operations    | `+ - * /`      |    |
| Comparison | Compare values     | `== != > <`    |    |
| Logical    | Combine conditions | `and or not`   |    |
| Assignment | Assign values      | `= += -=`      |    |
| Bitwise    | Binary operations  | `&             | ^` |
| Membership | Check in sequence  | `in`, `not in` |    |
| Identity   | Compare memory     | `is`, `is not` |    |

#### Comparison Operators :-

Comparison operators are used to compare two values.

They always return Boolean (True/False).

| Operator | Meaning          | Example  | Output  |
| -------- | ---------------- | -------- | ------- |
| `==`     | Equal to         | `5 == 5` | `True`  |
| `!=`     | Not equal        | `5 != 3` | `True`  |
| `>`      | Greater than     | `5 > 3`  | `True`  |
| `<`      | Less than        | `5 < 3`  | `False` |
| `>=`     | Greater or equal | `5 >= 5` | `True`  |
| `<=`     | Less or equal    | `3 <= 5` | `True`  |


```python
a = 10
b = 5
```

```python
print(a >b)
```

```python
print(a < b)
```

```python
print(a == b)
```

```python
print(a!=b)
```

```python
print(a>=b)
```

```python
print(a<=b)
```

```python
a = 'apple'
b = 'banana'
```

```python
print(a < b)
print(a > b)
print(a == b)
print(a != b)
print(a >= b)
print(a <= b)
```

##### Python compares strings character by character (lexicographically) using Unicode values. Strings are compared like a dictionary order.

**Important Rule :-** Comparison is based on:

First different character
    
ASCII/Unicode value

```python
print("a" < "b")   # True
print("A" < "a")   # True (uppercase < lowercase)
```

#### Logical Operators :-

Logical operators are used to combine multiple conditions.

Output is always Boolean(True/False).

| Operator | Meaning                      | Example          | Result |
| -------- | ---------------------------- | ---------------- | ------ |
| `and`    | Both conditions must be True | `True and False` | False  |
| `or`     | At least one condition True  | `True or False`  | True   |
| `not`    | Reverse the result           | `not True`       | False  |

**Simple Trick:-**

and = BOTH (ALL)

or = ANY

not = OPPOSITE


```python
a = 10
b = 5
```

```python
print(a>b and b<10)
```

```python
print(a>5 or b>10)
```

```python
print(not(a>8))
```

#### Bitwise Operators :-

Bitwise operators work on binary (bits : 0 & 1).

Used in low-level programming, optimization and interviews.

| Operator | Name        | Example  | Meaning        |
| -------- | ----------- | -------- | -------------- |
| `&`      | AND         | `5 & 3`  | Common bits    |
| `\|`     | OR          | `5 \| 3` | Any bit 1      |
| `^`      | XOR         | `5 ^ 3`  | Different bits |
| `~`      | NOT         | `~5`     | Invert bits    |
| `<<`     | Left Shift  | `5 << 1` | Shift left     |
| `>>`     | Right Shift | `5 >> 1` | Shift right    |

**Simple Tricks**

& → common bits

| → combine bits

^ → different bits

<< → ×2  ---> a << n = a × (2^n)
>> -> ÷2 .

~n → -(n+1)


```python
a = 6 # 110
b = 2 # 010
```

```python
print(a & b)
```

```python
print(a | b)
```

```python
print(a ^ b)
```

```python
print(~5)
```

```python
print(~a)
```

```python
print(a << b)
```

```python
print(5 << 1)
```

```python
print(a << 3)
```

```python
print(a >> b)
```

#### Membership Operators :-

Memebership operators are used to check wheather a value exits in a sequence or collection.

| Operator | Meaning                              |
| -------- | ------------------------------------ |
| `in`     | Returns True if value exists         |
| `not in` | Returns True if value does NOT exist |

**Shortcut Trick:-**

in → present

not in → not present


##### in operator -> check if elements is present

```python
a = 'apple'
print('a' in a)
```

```python
print('n' in a)
```

```python
print('le' in a)
```

##### not in operator -> checks if elements is NOT present.

```python
a = 'banana'
print('z' not in a)
```

```python
print('b' not in a)
```

```python
print('ana' not in a)
```

#### Identify Operators :-

Identify operators check whether two variable refer to the same object (same memory location).

| Operator | Meaning                                |
| -------- | -------------------------------------- |
| `is`     | True if both variables are same object |
| `is not` | True if both are different objects     |

**Simple Trick:-**

is → Checks same object (memory)

== → Checks values are equal


##### is operator -> checks same memory (same object).

```python
a = [1,2,3]
b = a
```

```python
print(a is b)
```

```python
b = [1,2,5]
print(a == b)
```

```python
a = [1,2,3]
b = (1,2,3)
print(a is b)
```

```python
a = [1,2,3]
b = [1,2]
print(a is b)
```

```python
a = [1,2,3]
b = [1,2,3]
print(a is b)
```

##### is not operator -> checks different memory (different objects).

```python
a = [1,2,3]
b = [1,2,3]
print(a is not b)
```

```python
a = [1,2]
b = {1,2}
print(a is not b)
```

##### == vs is

```python
a = [1,2,3]
b = [1,2,3]
print(a == b)
print(a is b)
```

```python
a = [1,2]
b = a
print(a == b)
print(a is b)
```
