#### Precedence Rule :-

Probably mean precedence rule (priority of operators).

It decides which operation runs first.

Hight ----->  Low

| Priority | Operators          |   |
| -------- | ------------------ | - |
| 1️⃣      | `()` (Parentheses) |   |
| 2️⃣      | `**` (Power)       |   |
| 3️⃣      | `* / // %`         |   |
| 4️⃣      | `+ -`              |   |
| 5️⃣      | `<< >>`            |   |
| 6️⃣      | `&`                |   |
| 7️⃣      | `^`                |   |
| 8️⃣      | `                  | ` |
| 9️⃣      | `== != > < >= <=`  |   |
| 🔟       | `not`              |   |
| 1️⃣1️⃣   | `and`              |   |
| 1️⃣2️⃣   | `or`               |   |


```python
print(2+3*4> 10 and not False)
```

```python
print(5 > 3 or 2 > 10 and not False)
```

```python
print(True or False and False)
```

```python
print((True or False) and False)
```

```python
a = 10
b = 20
print(a if a > b else b)
```

```python
print(False or True and not False)
```

```python
print(3 + 4 * 2 > 10 and not (5 < 2))
```
