### Set :-

A set is a collection of unique elements.

No duplocates allowed

unordered (no index)

Mutable (can change)


```python
a = {1,2,3}
b = {'apple','banana','cherry'}
c = {1,2,3,'a','b'}
print(a)
print(b)
print(c)
```

```python
a = {1,2,2,1,3,4}
print(a)
```

### Types :-

| Method             | Description                          | Example                            |
| ------------------ | ------------------------------------ | ---------------------------------- |
| `add(x)`           | Adds element                         | `a={1,2}; a.add(3) → {1,2,3}`      |
| `update(iterable)` | Adds multiple elements               | `a={1}; a.update([2,3]) → {1,2,3}` |
| `remove(x)`        | Removes element (error if not found) | `a={1,2}; a.remove(2) → {1}`       |
| `discard(x)`       | Removes element (no error)           | `a={1,2}; a.discard(5) → {1,2}`    |
| `pop()`            | Removes random element               | `a={1,2,3}; a.pop()`               |
| `clear()`          | Removes all elements                 | `a={1,2}; a.clear() → set()`       |
| `copy()`           | Returns copy of set                  | `b = a.copy()`                     |

| Method                   | Description            | Example                                  |
| ------------------------ | ---------------------- | ---------------------------------------- |
| `union()`                | Combines sets          | `a={1,2}; b={2,3}; a.union(b) → {1,2,3}` |
| `intersection()`         | Common elements        | `a.intersection(b) → {2}`                |
| `difference()`           | Elements in a not in b | `a.difference(b) → {1}`                  |
| `symmetric_difference()` | Non-common elements    | `a.symmetric_difference(b) → {1,3}`      |

| Method                          | Description                  | Example                            |
| ------------------------------- | ---------------------------- | ---------------------------------- |
| `intersection_update()`         | Updates with common elements | `a.intersection_update(b)`         |
| `difference_update()`           | Removes common elements      | `a.difference_update(b)`           |
| `symmetric_difference_update()` | Keeps non-common elements    | `a.symmetric_difference_update(b)` |

| Method         | Description        | Example           |
| -------------- | ------------------ | ----------------- |
| `issubset()`   | Checks if subset   | `a.issubset(b)`   |
| `issuperset()` | Checks if superset | `a.issuperset(b)` |
| `isdisjoint()` | No common elements | `a.isdisjoint(b)` |


```python
a = {1,2,3}
b = {3,4,5}

a.add(6)       # add element
a.update([7,8])      # add multiple elements

print(a.union(b))         
print(a.intersection(b))
print(a.difference(b))
print(a.symmetric_difference(b))
```
