## Python Data Types Comparison Table :-

| Feature           | String                 | List                | Tuple              | Set                    | Dictionary         |
| ----------------- | ---------------------- | ------------------- | ------------------ | ---------------------- | ------------------ |
| **Definition**    | Sequence of characters | Collection of items | Ordered collection | Unordered unique items | Key-value pairs    |
| **Syntax**        | `"abc"`                | `[1,2,3]`           | `(1,2,3)`          | `{1,2,3}`              | `{"a":1}`          |
| **Mutable**       | ❌ No                   | ✅ Yes               | ❌ No               | ✅ Yes                  | ✅ Yes              |
| **Ordered**       | ✅ Yes                  | ✅ Yes               | ✅ Yes              | ❌ No                   | ✅ (Py 3.7+)        |
| **Duplicates**    | ✅ Allowed              | ✅ Allowed           | ✅ Allowed          | ❌ Not allowed          | ❌ Keys not allowed |
| **Indexing**      | ✅ Yes                  | ✅ Yes               | ✅ Yes              | ❌ No                   | ❌ (use keys)       |
| **Slicing**       | ✅ Yes                  | ✅ Yes               | ✅ Yes              | ❌ No                   | ❌ No               |
| **Heterogeneous** | ❌ Mostly chars         | ✅ Yes               | ✅ Yes              | ✅ Yes                  | ✅ Yes              |
| **Dynamic size**  | ❌ Fixed                | ✅ Yes               | ❌ Fixed            | ✅ Yes                  | ✅ Yes              |
| **Access method** | Index                  | Index               | Index              | Loop only              | Key                |
| **Speed**         | Fast                   | Medium              | Faster than list   | Fast lookup            | Very fast lookup   |
| **Use case**      | Text data              | General collection  | Fixed data         | Unique items           | Structured data    |
