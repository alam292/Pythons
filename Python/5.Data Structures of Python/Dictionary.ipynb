### Dictionary :- 

A dictionary stores data in key:value pairs

Unordered 

Mutable

Keys must be unique



```python
d = {'name':'Rahul','age':13,'course':'math'}
print(d)
print(type(d))
print(len(d))
```

```python
d = {'a':1,'b':2,'c':3}
for key in d:
    print(key,d[key])
```

```python
student = {
    "name": "Rahul",
    "marks": {"math": 90, "eng": 85}
}

print(student["marks"]["math"])   # 90
```

```python
d = {"a": 1, "b": 2}

print("a" in d)   # True
```

### Types :-

| Method         | Description           | Example                      | Output          |
| -------------- | --------------------- | ---------------------------- | --------------- |
| `clear()`      | Remove all items      | `d.clear()`                  | `{}`            |
| `copy()`       | Returns copy          | `d2 = d.copy()`              | New dict        |
| `fromkeys()`   | Create dict from keys | `dict.fromkeys(['a','b'],0)` | `{'a':0,'b':0}` |
| `get()`        | Get value safely      | `d.get('a')`                 | Value / None    |
| `items()`      | All key-value pairs   | `d.items()`                  | dict_items      |
| `keys()`       | All keys              | `d.keys()`                   | dict_keys       |
| `values()`     | All values            | `d.values()`                 | dict_values     |
| `pop()`        | Remove key            | `d.pop('a')`                 | value           |
| `popitem()`    | Remove last item      | `d.popitem()`                | (key,value)     |
| `setdefault()` | Get or set default    | `d.setdefault('a',10)`       | value           |
| `update()`     | Update dict           | `d.update({'a':5})`          | updated dict    |
