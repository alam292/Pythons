#### Pyhton List methods :-

| Method               | Description              | Example                                     |
| -------------------- | ------------------------ | ------------------------------------------- |
| `append(x)`          | Adds element at end      | `a=[1,2]; a.append(3) → [1,2,3]`            |
| `extend(iterable)`   | Adds multiple elements   | `a=[1]; a.extend([2,3]) → [1,2,3]`          |
| `insert(i, x)`       | Insert at index          | `a=[1,3]; a.insert(1,2) → [1,2,3]`          |
| `remove(x)`          | Removes first occurrence | `a=[1,2,2]; a.remove(2) → [1,2]`            |
| `pop()`              | Removes last element     | `a=[1,2,3]; a.pop() → [1,2]`                |
| `pop(i)`             | Removes element at index | `a=[1,2,3]; a.pop(1) → [1,3]`               |
| `clear()`            | Removes all elements     | `a=[1,2]; a.clear() → []`                   |
| `index(x)`           | Returns index of element | `a=[1,2,3]; a.index(2) → 1`                 |
| `count(x)`           | Counts occurrences       | `a=[1,2,2]; a.count(2) → 2`                 |
| `sort()`             | Sorts list ascending     | `a=[3,1,2]; a.sort() → [1,2,3]`             |
| `sort(reverse=True)` | Sort descending          | `a=[1,2,3]; a.sort(reverse=True) → [3,2,1]` |
| `reverse()`          | Reverses list            | `a=[1,2,3]; a.reverse() → [3,2,1]`          |
| `copy()`             | Returns copy of list     | `a=[1,2]; b=a.copy()`                       |

##### count :-

##### sort :-

sort() is used to arrange list elements in order (ascending by default).

```python
a = [3,2,1]
a.sort()
print(a)
```

```python
a
```

```python
a.sort(reverse = True)
print(a)
```

```python
fruits = ['banana','mango','cherry','apple','date']
fruits.sort()
print(fruits)
```

```python
# case - sensitivity

fruits = ['banana','Mango','cherry','apple','Date']
fruits.sort()
print(fruits)
```

```python
a = ['apple','kiwi','date']
a.sort(key = len)
print(a)
```

##### copy :-

copy() is used to create a copy of a list.

It creates a new list with same elements

```python
a = [1,2,3]
b = a.copy()
print(a)
print(b)
```

```python
a = [1, 2, 3]
b = a.copy()

b.append(4)

print("a:", a)
print("b:", b)
```

##### reverse :-

reverse() is used to reverse the order of elements in a list.

It changes the list in-place (original list is modified).

```python
a = [1,2,3,4]
a.reverse()
print(a)
```

```python
fruits = ["apple", "banana", "mango"]
fruits.reverse()

print(fruits)
```

```python
a = [10, 20, 30]
x = a.reverse()

print(x)
```
