### Comprehension :-

A comprehension is a short and readable way to create collections(list,set,dict) in a single line.

It replaces loops with a compact syntax.

**OR**

Comprehension = Short way to write a loop

Instead of writing 4–5 lines, you write 1 line

#### Why Use Comprehension?

✔ Short & clean code

✔ Faster than loops

✔ Easy data transformation

✔ Widely used in real-world projects

**Syntax :- [expression for item in iterable if condition]**

##### Normal loop

```python
nums = [1,2,3,4]
square = []

for x in nums:
    square.append(x*x)

print(square)
```

##### comprehension

```python
nums = [1,2,3,4]
square = [x*x for x in nums]
print(square)
```

### List Comprehension 

##### Sqaure number

```python
nums = [1,2,3,4,5]
result = [x*x for x in nums]
print(result)
```

##### Even number

```python
nums = [1,2,3,4,5,6]
even = [x for x in nums if x%2==0]
print(even)
```

##### if-else

```python
nums = [1,2,3,4,5]
result = ['Even' if x%2==0 else 'Odd' for x in nums]
print(result)
```

### Dictionary Comprehension

```python
nums = [1,2,3,4,5]
d = {x : x*x for x in nums}
print(d)
```

### Set Comprehension 

```python
nums = [1,2,2,3,4]
s = {x for x in nums}
print(s)
```

### String Comprehension

**Syntax :- "".join([expression for char in string])**

"".join() converts list → string

```python
text = 'hello'
result = "".join([ch.upper() for ch in text])
print(result)
```

```python
# with condition

text = 'hello123'
result = "".join([ch for ch in text if ch.isalpha()])
print(result)
```

```python
# Replace character

text = 'banana'
result = "".join(['x' if ch=='a' else ch for ch in text])
print(result)
```
