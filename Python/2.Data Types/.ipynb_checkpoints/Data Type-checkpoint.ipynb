### Data Type :- 

A data type defines the type of value a variable can store.

Data Type = “Type of data” stored in a variable

```python
x = 10      # int data type
x
```

#### Quick Summary :- 

Numeric → Numbers

String → Text

List/Tuple → Ordered data

Set → Unique values

Dictionary → Key-value

#### Python Data Types Table :-

| Category      | Data Type | Example          | Description                         |
| ------------- | --------- | ---------------- | ----------------------------------- |
| 🔢 Numeric    | `int`     | `10`             | Whole numbers                       |
| 🔢 Numeric    | `float`   | `3.14`           | Decimal numbers                     |
| 🔢 Numeric    | `complex` | `2+3j`           | Complex numbers                     |
| 📝 Text       | `str`     | `"Hello"`        | Sequence of characters              |
| 📋 Sequence   | `list`    | `[1, 2, 3]`      | Ordered, mutable (changeable)       |
| 📦 Sequence   | `tuple`   | `(1, 2, 3)`      | Ordered, immutable (not changeable) |
| 🔁 Collection | `set`     | `{1, 2, 3}`      | Unordered, unique values            |
| 🔑 Mapping    | `dict`    | `{"a": 1}`       | Key-value pairs                     |
| ⚖️ Boolean    | `bool`    | `True` / `False` | Logical values                      |


##### 1.Numeric Data Types :-

Numerical data types are used to store number in Python.

Example - int -> (5,78) 

          float -> (39.4, 0.7)
          
          complex -> (2+5j) 

```python
a = 10 
b = 8
print(a+b)
```

```python
type(a+b)
```

```python
print(type(a))
```

```python
a = 8.87
b = 7.2
c = a+b
print(c)
```

```python
type(c)
```

```python
c = 2 + 3j
print(c)
print(type(c))
```

##### 2.Sequence Data Type :-

Sequence = an ordered collection of elements.
    
Elements are stored in a specific order (index-based)

Example :- str --> ("hello")

        list --> [1,2,3]

        tuple --> (1,2,3)

```python
text = 'python'
print(text)
```

```python
type(text)
```

```python
print(type(text))
```

```python
lis = [10,29,5]
print(list)
```

```python
print(type(lis))
```

```python
len(lis)
```

```python
tup = (28,8,2)
print(tup)
```

```python
print(type(tup))
```

```python
print(len(tup))
```

##### 3.Boolean Data Type :- 

Boolean is a data type that has only two values:

✅ True
❌ False

```python
a = True
b = False
print(a)
print(b)
```

```python
print(type(a))
```

```python
print(10>5)
print(19<7)
print(3==3)
print(7!=7)
```

##### 4.Collection Data Type :-

Collection = a group of multiple values stored in one variable.

Example :- list  , string, tuple , set , dictionary

##### 5.Mapping Data Type :-

Mapping = Key → Value pair data structure.
    
In Python, mapping is mainly done using dict (dictionary).

| Feature        | Description         |
| -------------- | ------------------- |
| Type           | `dict`              |
| Structure      | Key : Value         |
| Mutable        | ✅ Yes               |
| Ordered        | ✅ Yes (Python 3.7+) |
| Duplicate Keys | ❌ Not allowed       |


```python
student = {
    "name": "Ali",
    "age": 20,
    "course": "Python"}
print(student)
```

```python
type(student)
```

```python
len(student)
```
