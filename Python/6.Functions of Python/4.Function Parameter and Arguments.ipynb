### What are Parameters & Arguments ?

Parameter --> Variable in function definition.

Argument --> Actual value passed to function.

```python
def add(a,b):      # a,b ---> parameters
    return a + b

add (2,4)         # 2,4 ---> arguments
```

### Types of Arguments:-

### 1.Positonal Arguments :-

values are passed in order (position matters)

```python
def student(name, age):
    print(name, age)
    
print(student ('Ali',23))
print(student(22,'Ali'))
```

```python
# incorrect meaning order

student(22,'Ali')
```

### 2.Keyword Arguments :-

Arguments passed using paramter name.

order does NOT matter

```python
def student(name,age):
    print(name,age)
student(age = 24, name = 'Ali')
```

### 3.Default Arguments :- 

Provide default value if no argument is passed.

Default value is used only if argument not given.

```python
def greet(name = 'Guest'):
    print('Hello',name)

greet()
greet('Ali')
```

### 4.Variable-Length Arguments :-

Used when you don’t know how many inputs will come

##### A. *args (Non-keyword arguments):-

Accepts multiple values as tuple


```python
def total(*number):
    print(number)
    print(sum(number))

total(1,2,3,4)
```

##### B. **kwargs (Keyword arguments) :-

Accepts key-value pairs as dictionary.

```python
def info(**data):
    print(data)

info(name = 'Ali',age = 22)
```

```python
def order_food(item, quantity=1, *extras, **details):
    print("Item:", item)
    print("Quantity:", quantity)
    print("Extras:", extras)
    print("Details:", details)

order_food("Pizza", 2, "Cheese", "Olives", size="Large", drink="Coke")
```
