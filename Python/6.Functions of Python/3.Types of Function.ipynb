## Types of Functions :-

#### 1.Built-in Function :-

Built-in functions are pre-defined functions already available in Python.

You don't need to create them.

```python
print('hello')
```

```python
a = [1,2,3,4,5]
```

```python
print(len(a))
```

```python
print(sum(a))
```

```python
print(min(a))
```

```python
print(max(a))
```

```python
print(type(a))
```

### 2.User-defined Functions:-

Function created by the programmer using **def**.

```python
def add(a,b):
    return a+b
add(3,5)
```

```python
def even_odd(n):
    if n%2==0:
        return 'Even'
    else:
        return 'Odd'
print(even_odd(8))
```

### 3.Anonymous (Lambda) Function :-

A small , one-line function without a name.

Used with map(), filter().

**Syntax :- lambda arguments : expression**

```python
square = lambda x :x*x
print(square(4))
```

```python
# with map

num = [1,2,3]
result = list(map(lambda x : x*2, num))
print(result)
```

### Comparison Table

| Feature     | Built-in | User-defined | Lambda       |
| ----------- | -------- | ------------ | ------------ |
| Defined by  | Python   | User         | User         |
| Name        | Yes      | Yes          | No           |
| Size        | Small    | Any size     | One-line     |
| Reusability | Yes      | Yes          | Limited      |
| Example     | len()    | def add()    | lambda x:x+1 |


```python
# Built-in
print(len("Python"))

# User-defined
def greet(name):
    return "Hello " + name

print(greet("Ali"))

# Lambda
add = lambda a, b: a + b
print(add(2, 3))
```
