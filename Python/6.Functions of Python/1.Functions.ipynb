### Function :-

A function is a reusable block of code that performs a specific task.

Instead of writing the same code again and again , you define a function once and reuse it.


```python
def greet():
    print('hello, world!')
greet()
```

```python
def greet():
    print('Hello')
```

```python
greet()
```

### Why use functions? 

Code Reusability

Better readability

Easy  debugging

Modular programming

Avoid repetition (DRY Principle)
