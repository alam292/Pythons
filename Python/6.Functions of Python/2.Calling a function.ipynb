### What is Calling a function?

Calling a function means executing the function to run the code inside it .

You defined a function using **def** , but it only runs when you call it.

```python
# Example

def greet():
    print("Hello World")
greet()     # function call
```

### How it works (step-by-step):-

1.def greet() → Function is defined (stored in memory)

2.greet() → Function is called

3.Python executes the code inside the function

### Types of function Calling 

##### 1. Calling without Arguments

```python
def hello():
    print('Hi')

hello()
```

##### 2. Calling with Arguments

```python
def add(a,b):
    print(a+b)

add(5,8)
```

##### 3.Calling with Return value

```python
def square(x):
    return x*x

result = square(4)
print(result)
```

##### 4.Calling using Keyword Arguments

```python
def student(name,age):
    print(name, age)

student (age = 20, name = 'Alam')
```

##### 5.Calling Function Multiple Times 

```python
def greet():
    print('Hello')

greet()
greet()
greet()
```

#### Calling One Function Inside Another

```python
def add (a,b):
    return a+b

def display():
    result = add(2,3)
    print(result)

display()
```

#### Real-time example

```python
def login(user_name, password):
    if user_name == 'admin' and password == '12345':
        print('Login Success')
    else:
        print('Invalid')

login('admin','12345')
```
