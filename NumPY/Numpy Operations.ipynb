# NumPy Operations

### 1. Arithmetic Operations 

Numpy performs elements-wise operations on arrays

```python
import numpy as np
```

```python
a = np.array([10,20,30])
b = np.array([1,2,3])
```

```python
print(a+b)
```

```python
print(a-b)
```

```python
print(a*b)
```

```python
print(a/b)
```

```python
print(a%b)
```

### 2. Comparison Operations

used to compaare elements --> returns Boolean array (True/False)

```python
a = np.array([10,20,30,40])
```

```python
print(a>15)
```

```python
print(a<10)
```

```python
print(a=10)
```

```python
print(a==10)
```

```python
b = a%2==0
print(b)
```

##### used for filtering (Boolean Indexing)

```python
print(a[a>15])
```

```python
print(a>15)
```

### 3. Logical Operations

used to combine multiple conditions



```python
a = np.array([11,22,33,44])
```

```python
print((a>15) & (a<35))
```

##### using function

```python
a = np.logical_and(a>15,a<35)
print(a)
```

```python
b = np.logical_or(a>25 , a<15)
print(b)
```

```python
c = np.logical_not(a>20)
```
