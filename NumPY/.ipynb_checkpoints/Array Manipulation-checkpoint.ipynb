# Array Manipulation

### 1.Reshape Array (reshape()) 

change the shape (rows X columns) of an array without changing data.

**Key Points:-**

Total elements must match.

Use -1 to auto-calculate

```python
import numpy as np
```

```python
a = np.array([[1,2,3,4,5,6]])
print(a)
```

```python
a.ndim
```

```python
a.shape
```

```python
b = a.reshape((2,3))
print(b)
```

```python
c = a.reshape((6,1))
print(c)
```

```python
d = a.reshape((3,2))
print(d)
```

```python
e = a.reshape((3,3))
print(e)
```

### 2. Flatten Array (flatten() , ravel())

convert mutli-dimensional array ----> 1-D array

| Function    | Behavior                  |
| ----------- | ------------------------- |
| `flatten()` | Returns **copy**          |
| `ravel()`   | Returns **view (faster)** |


```python
a = np.array([[1,2,3],[4,5,6]])
print(a)
```

```python
a.ndim
```

```python
print(a.flatten())
```

```python
b = a.flatten()
print(b)
```

```python
b.ndim
```

```python
print(a.ravel())
```

### 3. Transpose (T)

convert rows ----> columns

```python
a = np.array([[1,2,3,4],[5,6,7,8]])
print(a)
```

```python
a.shape
```

```python
b = a.T
print(b)
```

```python
b.shape
```

### 4. Concatenation (concatenate())

Joins array along an axis

```python
a = np.array([[1,2,3]])
b = np.array([[11,22,33]])
```

```python
# Row-wise

print(np.concatenate((a,b),axis = 0))
```

```python
# Column-wise

print(np.concatenate((a.T,b.T),axis=1))
```

### 5. Splitting Array (split() & array_split())

breaks array into multiple parts

| Function        | Behavior         |
| --------------- | ---------------- |
| `split()`       | Equal parts only |
| `array_split()` | Unequal allowed  |


```python
a = np.arange(6)
```

```python
print(a)
```

```python
print(np.split(a,3))
```

```python
print(np.split(a,2))
```

```python
print(np.split(a,6))
```

### 6. Stacking (hstack(), vstack(),dstack())

stacks array in differnt directions

```python
a = np.array([1,2,3])
b = np.array([4,5,6])
```

```python
# horizontal

print(np.hstack((a,b)))
```

```python
# vertical

print(np.vstack((a,b)))
```

```python
# depth

print(np.dstack((a,b)))
```
