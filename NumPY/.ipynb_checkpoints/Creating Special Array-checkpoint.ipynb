# Creating Special Arrays

### 1. np.zeros() ---> Array filled with 0

creates an array where all elements are 0.

```python
import numpy as np
```

```python
a = np.zeros((2,4))
print(a)
```

```python
a
```

```python
type(a)
```

```python
len(a)
```

```python
a = np.zeros((5,9))
print(a)
```

### 2. np.ones() ---> Array filled with 1

create an array where all elements are 1.

```python
a = np.ones((3,2))
print(a)
```

```python
len(a)
```

```python
a = np.ones((1,9))
print(a)
```

```python
len(a)
```

### 3. np.full() ----> Custom value Array

creates an array fulled with a specific value.

```python
a = np.full((4,3),0.5)
print(a)
```

```python
a = np.full((8,4),5)
print(a)
```

### 4. np.arange() ----> Range of values

Works like python range() but returnws an array 

**Formate : start : stop : size step**

```python
a = np.arange(1,10,2)
print(a)
```

```python
a = np.arange(5,55,5)
print(a)
```

### 5. np.linspace()  ---> Equal spacing

creates numbers evenly spaced between two values.

```python
a = np.linspace(0,1,5)
print(a)
```

```python
a = np.linspace(1,10,20)
print(a)
```

```python
a = np.linspace(-10,10,20)
print(a)
```

```python
a = np.linspace(1,100,20)
print(a)
```

### 6 .np.eye()  ----> Identity Matrix

creates a martix with 1s on diagonal, 0s elsewhere

very important in linear algebra

```python
a = np.eye(4,4)
print(a)
```

```python
a = np.eye(4)
print(a)
```

```python
a = np.eye(2,4)
print(a)
```

### 7. np.random.rand() ---> Random values (0 to 1).    Uniform Distribution

generates random numbers between 0 and 1.

```python
import random
```

```python
a = np.random.rand(2,4)
print(a)
```

```python
a = np.random.rand(3,3)
print(a)
```

### 8. np.random.randint() ---> Random integers

generates random integers in a given range

```python
a = np.random.randint(1,10,(2,3))
print(a)
```

```python
a = np.random.randint(1,10,(2,3))
print(a)
```

```python
a = np.random.randint(1,10,(2,3))
print(a)
```

### 9. np.random.randn() ---> Random numbers (-infinity to +infinity) .     Normal Distribution (Gaussian)

randn() generates numbers from a standard normal distribution

**Syntax :- np.random.randn(d0 , d1, ......., dn)

```python
a = np.random.randn(5)
print(a)
```

```python
a = np.random.randn(5)
print(a)
```

```python
a = np.random.randn(10)
print(a)
```

```python
a = np.random.randn(100)
print(a)
```

```python
a = np.random.randn(500)
print(a)
```
