# Data Types Conversion in NumPy (astype())

### **Syntax :- array.astype(new_dtype)**

#### integer -----> float

```python
a = np.array([1,2,3])
new_a = a.astype(float)
print(new_a)
```

```python
print(new_a.dtype)
```

#### float ------> integer

```python
b = np.array([12.3, 67.3, 278.8])
b2 = b.astype(int)
print(b2)
```

```python
print(b2.dtype)
```

#### convert to string

```python
a = np.array([10,70,30])
a2 = a.astype(str)
print(a2)
```

```python
print(a2.dtype)
```

#### convert to boolean

```python
a = np.array([0,1,4,0,3,0])
a2 = a.astype(bool)
print(a2)
```

```python
print(a2.dtype)
```

```python
print(a2)
```

#### convert to specific types(int32, float32)

```python
a = np.array([1,2,3,4])
print(a)
print(a.dtype)
```

```python
a2 = a.astype(np.float32)
print(a2)
print(a2.dtype)
```
